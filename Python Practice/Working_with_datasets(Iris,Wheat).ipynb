4. Higher Dimensionality Visualizations- Parallel Coordinates

```python
from sklearn.datasets import load_iris
from pandas.plotting import parallel_coordinates

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

# Look pretty...
matplotlib.style.use('ggplot')

# Load up SKLearn's Iris Dataset into a Pandas DataFrame
data = load_iris() # <class 'sklearn.utils.Bunch'>
df = pd.DataFrame(data.data, columns=data.feature_names)

df['target_names'] = [data.target_names[i] for i in data.target]

# Parallel Coordinates Start Here:
plt.figure()
parallel_coordinates(df, 'target_names')
plt.show()
```

1. Histograms

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the seeds dataset into a dataframe
seeds_df = pd.read_csv('C:/Users/Suhaab Shah/Desktop/Wheat.csv')

# 2. Create a slice of the dataframe that only includes the 'area' and 'perimeter' features
area_perimeter_slice = seeds_df[['area', 'perimeter']]

# 3. Create another slice that only includes the 'groove' and 'asymmetry' features
groove_asymmetry_slice = seeds_df[['groove', 'asymmetry']]

# 4. Create a histogram for the 'area' and 'perimeter' slice
area_perimeter_slice.hist(alpha=0.75)
plt.xlabel('Area and Perimeter')
plt.ylabel('Frequency')
plt.title('Histogram of Area and Perimeter')
plt.show()

# Create a histogram for the 'groove' and 'asymmetry' slice
groove_asymmetry_slice.hist(alpha=0.75)
plt.xlabel('Groove and Asymmetry')
plt.ylabel('Frequency')
plt.title('Histogram of Groove and Asymmetry')
plt.show()
```

2. 2D Scatter Plots

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the seeds dataset into a dataframe
data = pd.read_csv("C:/Users/Suhaab Shah/Desktop/Wheat.csv")

# Create a 2D scatter plot for area and perimeter features
plt.scatter(data['area'], data['perimeter'])
plt.xlabel('Area')
plt.ylabel('Perimeter')
plt.title('Scatter Plot: Area vs Perimeter')
plt.show()

# Create a 2D scatter plot for groove and asymmetry features
plt.scatter(data['groove'], data['asymmetry'])
plt.xlabel('Groove')
plt.ylabel('Asymmetry')
plt.title('Scatter Plot: Groove vs Asymmetry')
plt.show()

# Create a 2D scatter plot for compactness and width features
plt.scatter(data['compactness'], data['width'])
plt.xlabel('Compactness')
plt.ylabel('Width')
plt.title('Scatter Plot: Compactness vs Width')
plt.show()
```

3. 3D Scatter Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the seeds dataset into a dataframe
data = pd.read_csv("C:/Users/Suhaab Shah/Desktop/Wheat.csv")

# Create a 3D scatter plot for area, perimeter, and asymmetry features
fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')
ax1.scatter(data['area'], data['perimeter'], data['asymmetry'], c='red')
ax1.set_xlabel('Area')
ax1.set_ylabel('Perimeter')
ax1.set_zlabel('Asymmetry')
ax1.set_title('3D Scatter Plot: Area, Perimeter, and Asymmetry')
plt.show()

# Create a 3D scatter plot for width, groove, and length features
fig = plt.figure()
ax2 = fig.add_subplot(111, projection='3d')
ax2.scatter(data['width'], data['groove'], data['length'], c='green')
ax2.set_xlabel('Width')
ax2.set_ylabel('Groove')
ax2.set_zlabel('Length')
ax2.set_title('3D Scatter Plot: Width, Groove, and Length')
plt.show()
```

4. Parallel Coordinates

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates

# Load the seeds dataset into a dataframe
data = pd.read_csv("C:/Users/Suhaab Shah/Desktop/Wheat.csv")

# Drop the 'area' and 'perimeter' features from the dataframe
data = data.drop(['area', 'perimeter'], axis=1)

# Plot a parallel coordinates chart, grouped by the 'wheat_type' feature
plt.figure(figsize=(10, 6))
parallel_coordinates(data, 'wheat_type', color=['blue', 'green', 'red'], alpha=0.4)
plt.title('Parallel Coordinates Chart')
plt.xlabel('Features')
plt.ylabel('Values')
plt.legend(loc='upper right')
plt.show()
```

5. Andrew’s Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import andrews_curves

# Load the seeds dataset into a dataframe
data = pd.read_csv("C:/Users/Suhaab Shah/Desktop/Wheat.csv")

# Plot an Andrew's Curve chart, grouped by the 'wheat_type' feature
plt.figure(figsize=(10, 6))
andrews_curves(data, 'wheat_type', color=['blue', 'green', 'red'], alpha=0.4)
plt.title("Andrew's Curve Chart")
plt.xlabel('Features')
plt.ylabel('Values')
plt.legend(loc='upper right')
plt.show()
```
